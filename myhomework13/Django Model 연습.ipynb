# Conditions(OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10)
```

```python
from django.db.models import Q
```

```python
Q(pk__gte=5)
```

```python
Q(pk_lt=10)
```

```python
Q(pk__gte=5) & Q(pk__lt=10)
```

```python
Q(pk__gte=5) | Q(pk__lt=10)
```

```python
qs = Shop.objects.all()
qs.filter(
    Q(pk__gte=5) | Q(pk__lt=10),
)
```

```python
conditions = Q(pk__gte=5)
conditions |= Q(pk__lt=10)

conditions
```

# Update Rows

### 지정 Row

```python
qs = Shop.objects.all()
shop = qs.get(pk=10)
```

```python
shop.name, shop.telephone
```

```python
shop.telephone = '042-1234-0006'
```

```python
shop.save()
```

### 지정 범위의 Rows

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
qs.update(
    description='not used description'
)
```

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
for shop in qs:
    shop.description = 'now used description'
    shop.save()
```
