# Comditions(OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10)   # AND
```

```python
from django.db.models import Q
```

```python
Q(pk__gte=5)
```

```python
Q(pk__lt=10)
```

```python
Q(pk__gte=5) & Q(pk__lt=10)
```

```python
Q(pk__gte=5) | Q(pk__lt=10)
```

```python
qs = Shop.objects.all()
qs.filter(
    Q(pk__gte=5) | Q(pk__lt=10)
)
```

```python
conditions=Q(pk__get=5)
# conditions=conditions | Q(pk__lt=10)
conditions |= Q(pk__lt=10)

conditions
```

## 지정 범위의 Row

```python
qs = Shop.objects.all()

qs = qs.filter(pk__lte=10)
qs.update(
    description="not used discription"
)
```

```python
qs = Shop.objects.all()

qs = qs.filter(pk__lte=10)

for shop in qs:
    shop.description="not used discription"
    shop.save()
```
